```python
diry = 'data'
sub_diry = 'outages'
new_dir = 'geojson'
# get list of outage files
json_files = []
needle = 'outage'
for n in os.listdir(os.path.join(diry, sub_diry)):
    if n in new_files: # make sure it's an outage file
        json_files.append(n)
print(json_files) # see how it loooks
print(len(json_files) ) # should be 42
```

```python
for fn in json_files:
    with open( os.path.join(diry, fn), 'r') as f:
        dirty_data = json.loads(f.read())

    features = []

    for d in dirty_data:
        try:
            d['outageDevices']
            pts = []
            for pair in d['outageDevices']:
                lat_long = (float(pair['longitude'] ), float( pair['latitude']))

                pts.append(lat_long)
                print( lat_long)
            end = (float(d['outageDevices'][0]['longitude'] ), float( d['outageDevices'][0]['latitude']))
            pts.append(end)
            poly = Polygon([pts])
            features.append(Feature(geometry=poly))
        except KeyError:
            print(f" no outageDevices for{d['outageNumber']}")
    feature_collection = FeatureCollection(features)
    new_path = os.path.join(new_dir, fn[:-4] + 'geojson')
    with open(new_path, 'w') as f:
        dump(feature_collection, f)
```
